```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
API_KEY = ''
```

```python
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol=IBM&apikey=demo'
r = requests.get(url)
data = r.json()
```

```python
time = data['Time Series (Daily)']
df = pd.DataFrame.from_dict(time, orient='index',dtype=float)
df
```

```python
df.columns = ['open', 'high', 'low', 'close', 'adjusted close', 'volume', 'dividend amount', 'split coefficient']
df
```

```python
df.index.name = 'date'
df
```

```python
# plot the open and close values
fig, ax = plt.subplots(figsize=(10,8))
ax.plot(df['open'], label='open')
ax.plot(df['close'], label='close')
ax.legend()
plt.show()
```
